```python
import pandas as pd
import numpy as np
# import array
```

```python
# Load data from csv and read into pandas
csv_path = "Resources/purchase_data.csv"
purchaseData_df = pd.read_csv(csv_path)
```

```python
purchaseData_df.head()
```

```python
# get total revenue & number of transactions
revenue = purchaseData_df["Price"].sum()
transactionCount = len(purchaseData_df)

#revenue
#transactionCount
```

```python
# get list of unique players and display total

uniquePlayers = purchaseData_df["SN"].unique()
uniquePlayersTot = len(uniquePlayers)
print(f"The total number of players is {uniquePlayersTot}")

```

```python
# convert list of unique players into DataFrame for pandas
uniquePlayers_df = pd.DataFrame({
    "SN": uniquePlayers
    })
# uniquePlayers_df.count()
# uniquePlayers_df.head
```

```python
# get list of unique items and display total

uniqueItems = purchaseData_df["Item ID"].unique()
uniqueItemsTot = len(uniqueItems)
print(f"The total number of items is {uniqueItemsTot}")
```

```python
# convert list of unique items into DataFrame for pandas

uniqueItems_df = pd.DataFrame({
    "Item ID": uniqueItems
})

# uniqueItems_df.head()
# print(uniqueItems_df.count())


# create DataFrame of all unique items with prices
itemPrice_df = purchaseData_df[["Item ID", "Price"]]
itemPrice_df.sort_values("Item ID", inplace=True)
itemPrice_df.drop_duplicates("Item ID", keep="first", inplace=True)
# itemPrice_df

# len(itemPrice_df)
# itemPrice_df.head()
```

```python
# get average purchase price for all transactions

purchaseData_df["Price"] = purchaseData_df["Price"].astype(float)

# purchaseData_df.dtypes

avgPricePurchases = purchaseData_df["Price"].mean()

# get average item price

avgPriceItems = itemPrice_df["Price"].mean()

```

```python
purchasingAnalysis_df = pd.DataFrame({
    "No. Unique items" : uniqueItemsTot,
    "Avg Price" : avgPricePurchases,
    "No. of Purchaes" : transactionCount,
    "Total Revenue" : revenue
}, index = [0])

purchasingAnalysis_df["Avg Price"] = purchasingAnalysis_df["Avg Price"].map("${:.2f}".format)
purchasingAnalysis_df["Total Revenue"] = purchasingAnalysis_df["Total Revenue"].map("${:.2f}".format)

purchasingAnalysis_df.head()
```

```python
# populate variables for Gender Demographics

# xxPlayerPurchases will create data frame with all purchases from specific gender
# xxPlayers_list will create list of all unique players from specific gender
# xxPlayersTot is number of players with specific gender

malePlayerPurchases_df = purchaseData_df.loc[(purchaseData_df["Gender"] == "Male"), :]
malePlayerPurchases_df

malePlayers_df = purchaseData_df.loc[(purchaseData_df["Gender"] == "Male"), ["SN", "Gender"]]
malePlayers_list = malePlayers_df["SN"].unique()
malePlayersTot = len(malePlayers_list)
# malePlayersTot

malePlayerShare = malePlayersTot / uniquePlayersTot


femalePlayerPurchases_df = purchaseData_df.loc[(purchaseData_df["Gender"] == "Female"), :]
#femalePlayerPurchases_df

femalePlayers_df = purchaseData_df.loc[(purchaseData_df["Gender"] == "Female"), ["SN", "Gender"]]
femalePlayers_list = femalePlayers_df["SN"].unique()
femalePlayersTot = len(femalePlayers_list)
# femalePlayersTot

femalePlayerShare = femalePlayersTot / uniquePlayersTot

otherPlayerPurchases_df = purchaseData_df.loc[(purchaseData_df["Gender"] == "Other / Non-Disclosed"), :]
# otherPlayerPurchases_df

otherPlayers_df = purchaseData_df.loc[(purchaseData_df["Gender"] == "Other / Non-Disclosed"), ["SN", "Gender"]]
otherPlayers_list = otherPlayers_df["SN"].unique()
otherPlayersTot = len(otherPlayers_list)
# otherPlayersTot

otherPlayerShare = otherPlayersTot / uniquePlayersTot
```

```python
# generate data frame and format Gender Demographics

playerDemographics_df = pd.DataFrame({
    "Total Players" : [malePlayersTot, femalePlayersTot, otherPlayersTot],
    "Percent of Players" : [malePlayerShare * 100, femalePlayerShare * 100, otherPlayerShare * 100],
}, index = ["Male", "Female", "Other / Non-Disclosed"])


playerDemographics_df["Percent of Players"] = playerDemographics_df["Percent of Players"].map("{:.2f}%".format)

playerDemographics_df.head()
```

```python
# populate variables for Purchase Analysis (Gender) data frame

malePurchaseCount = malePlayerPurchases_df["SN"].count()
malePurchaseTot = malePlayerPurchases_df["Price"].sum()
avgMalePurchase = round(malePurchaseTot / malePurchaseCount, 2)
avgMalePurchasePerson = round(malePurchaseTot / malePlayersTot, 2)

femalePurchaseCount = femalePlayerPurchases_df["SN"].count()
femalePurchaseTot = femalePlayerPurchases_df["Price"].sum()
avgFemalePurchase = round(femalePurchaseTot / femalePurchaseCount, 2)
avgFemalePurchasePerson = round(femalePurchaseTot / femalePlayersTot, 2)

otherPurchaseCount = otherPlayerPurchases_df["SN"].count()
otherPurchaseTot = otherPlayerPurchases_df["Price"].sum()
avgOtherPurchase = round(otherPurchaseTot / otherPurchaseCount, 2)
avgOtherPurchasePerson = round(otherPurchaseTot / otherPlayersTot, 2)

# print(malePurchaseCount)
# print(malePurchaseTot)
# print(avgMalePurchase)
# print(avgMalePurchasePerson)

# print(femalePurchaseCount)
# print(femalePurchaseTot)
# print(avgFemalePurchase)
# print(avgFemalePurchasePerson)

# print(otherPurchaseCount)
# print(otherPurchaseTot)
# print(avgOtherPurchase)
# print(avgOtherPurchasePerson)
```

```python
# generate data frame and format for Purchase Analysis (Gender)

genderPurchasingAnalysis_df = pd.DataFrame({
    "Purchase Count" : [femalePurchaseCount, malePurchaseCount, otherPurchaseCount],
    "Avg Purchase Price" : [avgFemalePurchase, avgMalePurchase, avgOtherPurchase],
    "Tot Purchase Value" : [femalePurchaseTot, malePurchaseTot, otherPurchaseTot],
    "Avg Tot Purchase / Person" : [avgFemalePurchasePerson, avgMalePurchasePerson, avgOtherPurchasePerson]
}, index = ["Female", "Male", "Other"])

genderPurchasingAnalysis_df["Avg Purchase Price"] = genderPurchasingAnalysis_df["Avg Purchase Price"].map("${:.2f}".format)
genderPurchasingAnalysis_df["Tot Purchase Value"] = genderPurchasingAnalysis_df["Tot Purchase Value"].map("${:.2f}".format)
genderPurchasingAnalysis_df["Avg Tot Purchase / Person"] = genderPurchasingAnalysis_df["Avg Tot Purchase / Person"].map("${:.2f}".format)

genderPurchasingAnalysis_df
```

```python
# generate bins and data frame for Age Demographics

bins = [0, 9, 14, 19, 24, 29, 34, 39, 1000]
binLabels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchaseData_df["Age Grp"] = pd.cut(purchaseData_df["Age"], bins, labels = binLabels)
purchaseData_df.head()

uniquePlayerData_df = purchaseData_df[["SN", "Age", "Gender", "Age Grp"]]
uniquePlayerData_df.drop_duplicates("SN", keep="first", inplace=True)


playerAgeBinCount = uniquePlayerData_df.groupby("Age Grp")
playerAgeBinCount = playerAgeBinCount["Age Grp"].count().values
lovPlayerAgeBinCount = playerAgeBinCount.tolist()
lovPlayerAgeBinCount

playerAgeBinPercentage = []

for number in lovPlayerAgeBinCount:
        playerAgeBinPercentage.append(number / uniquePlayersTot * 100)

ageDemo = zip(lovPlayerAgeBinCount, playerAgeBinPercentage)

     
ageDemo_df = pd.DataFrame(ageDemo,
                          columns=["Total Count", "Percent of Players"],
                          index = [binLabels])

ageDemo_df["Percent of Players"] = ageDemo_df["Percent of Players"].map("{:.2f}%".format)

ageDemo_df


# purchaseData_df.head()
# uniquePlayerData_df.head()
# len(uniquePlayerData_df)
# playerAgeBins["Age Grp"].count()
# playerAgeBinPercentage
```

```python
# generate data frame for Purchasing Analysis (Age)

# purchaseData = purchaseData_df.groupby("Age Grp")

purchaseAnalysis_PC = purchaseData_df[["SN", "Age Grp"]]
purchaseAnalysis_PC = purchaseAnalysis_PC.groupby("Age Grp").count()
purchaseAnalysis_PC

purchaseAnalysis_APP = purchaseData_df[["Price", "Age Grp"]]
purchaseAnalysis_APP = purchaseAnalysis_APP.groupby("Age Grp").mean()
purchaseAnalysis_APP

purchaseAnalysis_TPV = purchaseData_df[["Price", "Age Grp"]]
purchaseAnalysis_TPV = purchaseAnalysis_TPV.groupby("Age Grp").sum()
purchaseAnalysis_TPV

purchaseDataPP = purchaseData_df[["SN","Price"]]
purchaseDataPP = purchaseDataPP.groupby(["SN"]).mean()
purchaseDataPP = pd.merge(purchaseDataPP, uniquePlayerData_df, on="SN", how="left")
purchaseDataPP = purchaseDataPP.groupby("Age Grp").mean()
purchaseDataPP["Price"]

purchaseAnalysis_combined = pd.concat([purchaseAnalysis_PC, purchaseAnalysis_APP, purchaseAnalysis_TPV, purchaseDataPP["Price"]], 
                                      axis=1, 
                                      join="inner",
                                     )

purchaseAnalysis_combined.columns = ["Purchase Count", "Avg Purchase Price", "Tot Purchase Value", "Avg Tot Purchase / Person"]

purchaseAnalysis_combined["Avg Tot Purchase / Person"] = purchaseAnalysis_combined["Avg Tot Purchase / Person"].map("${:.2f}".format)

purchaseAnalysis_combined
```

```python
purchaseData_PCPP = purchaseData_df.groupby("SN").count()
purchaseData_PCPP = purchaseData_PCPP["Purchase ID"]
purchaseData_PCPP = purchaseData_PCPP.sort_values(ascending=False)
                                             
# purchaseData_df

purchaseData_TPVPP = purchaseData_df.groupby("SN").sum()
purchaseData_TPVPP = purchaseData_TPVPP["Price"]
purchaseData_TPVPP = purchaseData_TPVPP.sort_values(ascending=False)
purchaseData_TPVPP

purchaseDataPP = pd.concat([purchaseData_PCPP, purchaseData_TPVPP],
                          axis=1,
                          join="inner",
                          )

purchaseDataPP.columns = ["Purchase Count", "Tot Purchase Value"]
purchaseDataPP["Avg Purchase Price"] = purchaseDataPP["Tot Purchase Value"] / purchaseDataPP["Purchase Count"]

purchaseDataPP = purchaseDataPP.sort_values(["Tot Purchase Value"], ascending=False)

purchaseDataPP["Tot Purchase Value"] = purchaseDataPP["Tot Purchase Value"].map("${:.2f}".format)
purchaseDataPP["Avg Purchase Price"] = purchaseDataPP["Avg Purchase Price"].map("${:.2f}".format)

purchaseDataPP.head(5)
```
